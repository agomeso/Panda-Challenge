## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create a simplified series with SN, Gender and Age and remove the duplicates
players = purchase_data_df[["SN","Gender","Age"]].drop_duplicates()

#Calculate a series of total counts by gender
players_df = players["Gender"].value_counts()

#Create a list of percentages (count/total)
pct_player = players["Gender"].value_counts()/players["Gender"].count()*100

# #Put it all together in one df
gender_df = pd.DataFrame({"Total Count":players_df,"Percentage of Players":pct_player})

#Format the percetage
gender_df["Percentage of Players"]= gender_df["Percentage of Players"].map("{:.2f}%".format)

#Print
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate purchase count series
gender_ct = purchase_data_df["Gender"].value_counts()

#Calculate average purchase price by gender
group_by_gender = purchase_data_df.groupby(["Gender"])
mean_by_gender = group_by_gender.mean().head()
price_by_gender = mean_by_gender["Price"]

#Sum the purchase value by gender
sum_by_gender = group_by_gender.sum().head()
value = sum_by_gender["Price"]

#Calculate average total purchase per person by gender
group_by_person = purchase_data_df.groupby(["Gender","SN"])
purch_by_person = group_by_person.sum()
purch_by_person_and_gender = purch_by_person.groupby(["Gender"])
purch_by_person_and_gender = purch_by_person_and_gender.mean()
ave_total_pp = purch_by_person_and_gender["Price"]

#Create the df
purchasing_analysys_gender = pd.DataFrame(
    {"Purchase Count":gender_ct,
     "Average Purchase Price":price_by_gender,
     "Total Purchase Value":value,
    "Avg Total Purchase per Person":ave_total_pp})

#Format it
purchasing_analysys_gender["Average Purchase Price"]= purchasing_analysys_gender["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysys_gender["Total Purchase Value"]= purchasing_analysys_gender["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysys_gender["Avg Total Purchase per Person"]= purchasing_analysys_gender["Avg Total Purchase per Person"].map("${:.2f}".format)

#Print it
purchasing_analysys_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins
bins = [0,9.9,14,19,24,29,34,39,100]

# #Create names for the bins
names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

purchase_data_df["Age Ranges"]=pd.cut(purchase_data_df["Age"],bins,labels=names,include_lowest=True)
purchase_data_df

#spot check the data
# purchase_data_df.to_csv("check_this.csv", index=False, header=True)

#Calculate purchase count, average price and total purchase value
PA_df = purchase_data_df.groupby("Age Ranges")
ct_by_age = PA_df["Purchase ID"].count()
ave_price_by_age = PA_df["Price"].mean()
total_pur_val_by_age = PA_df["Price"].sum()

#Calculate average total purchase per person and age
group_by_ppl = purchase_data_df.groupby(["Age Ranges","SN"])
purch_by_ppl = group_by_ppl.sum()
purch_by_ppl
purch_by_person_and_age = purch_by_ppl.groupby(["Age Ranges"])
purch_by_person_and_age = purch_by_person_and_age.mean()
ave_total_pp_age = purch_by_person_and_age["Price"]

#Create the df
purchasing_analysys_age = pd.DataFrame(
    {"Purchase Count":ct_by_age,
     "Average Purchase Price":ave_price_by_age,
     "Total Purchase Value":total_pur_val_by_age,
    "Avg Total Purchase per Person":ave_total_pp_age})

#Format it
purchasing_analysys_age["Average Purchase Price"]= purchasing_analysys_age["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysys_age["Total Purchase Value"]= purchasing_analysys_age["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysys_age["Avg Total Purchase per Person"]= purchasing_analysys_age["Avg Total Purchase per Person"].map("${:.2f}".format)

#Print it
purchasing_analysys_age
```
